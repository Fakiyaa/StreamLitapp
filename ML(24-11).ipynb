```python
df=df.rename(columns={'Sex':'Gender','Blood Pressure':'Blood_Pressure','Heart Rate':'Heart_Rate','Family History':'Family_History','Previous Heart Problems':'Previous_Heart_Problems','Stress Level':'Stress_Level','Heart Attack Risk':'Heart_Attack_Risk'})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoded = encoder.fit_transform(df[['Gender']]).toarray()
encoded
```

```python
cols = encoder.get_feature_names_out(['Gender'])
cols
```

```python
encoded_df = pd.DataFrame(encoded, columns=cols)
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df
```

```python
df=df[['Age','Cholesterol','Blood_Pressure','Heart_Rate', 'Diabetes','Family_History','Smoking', 'Previous_Heart_Problems','Stress_Level', 'Heart_Attack_Risk','Gender_Female','Gender_Male']].copy() 
```

```python
df
```

```python
df['Gender_Female']=df.Gender_Female.astype('int')
```

```python
df['Gender_Male']=df.Gender_Male.astype('int')
```

```python
df
```

```python
df[['High_Blood_Pressure', 'Low_Blood_Pressure']] = df['Blood_Pressure'].str.split('/', expand=True)
```

```python
df
```

```python
df.drop('Blood_Pressure', axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=linear_model.LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['Age','Cholesterol','Heart_Rate', 'Diabetes','Family_History','Smoking', 'Previous_Heart_Problems','Stress_Level', 'Gender_Female','Gender_Male','High_Blood_Pressure','Low_Blood_Pressure']],df['Heart_Attack_Risk'],test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```
